__A kiválasztás célja:__ A klaszterezéshez a __napi játékidő (daily_gaming_hours)__ és az __alvásidő (sleep_hours)__ attribútumokat választottam. A célom az volt, hogy jól elkülöníthető életmódbeli profilokat hozzak létre. Arra kerestem a választ, hogy a játékkal töltött idő mennyisége milyen mintázatokat mutat a pihenésre fordított idővel, és hogy a "sokat játszó" egyének automatikusan az alvásból vesznek-e el időt, vagy léteznek ettől eltérő, sajátos csoportok is.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN


X = df[['daily_gaming_hours', 'sleep_hours']].dropna()


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python

err = []

for i in range(1, 11):
   
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10) 
    kmeans.fit(X_scaled)
    err.append([kmeans.n_clusters, kmeans.inertia_])


err = np.asarray(err) 


plt.figure(figsize=(8, 5))
sns.lineplot(x=err[:,0], y=err[:,1], marker='o', color='b')
sns.scatterplot(x=err[:,0], y=err[:,1], color='red', s=80, zorder=5)

plt.title('Könyök módszer az optimális klaszterszám (K) meghatározásához')
plt.xlabel('Klaszterek száma (K)')
plt.ylabel('Inercia (Klaszteren belüli hiba)')
plt.xticks(range(1, 11))
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

A görbén látszik, hogy a hiba meredek csökkenése a __K=3__ értéknél törik meg és kezd ellaposodni. 
A K-Means algoritmusnál a __3 klaszter__ (játékosprofil) beállítása  az optimális választás, ennél több csoport kialakítása már nem csökkentené érdemben a hibát.

```python

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_scaled)


plt.figure(figsize=(8, 6))

sns.scatterplot(x=X['daily_gaming_hours'], y=X['sleep_hours'], hue=kmeans_labels, palette='viridis', s=60)


centers = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroidok')

plt.title('K-Means Klaszterezés (k=3): Játékidő vs. Alvásidő')
plt.xlabel('Napi játékidő (óra)')
plt.ylabel('Alvásidő (óra)')
plt.legend(title='Klaszter')
plt.show()
```

A K-Means algoritmus három jól értelmezhető játékosprofilt különített el: a sokat játszó, __átlagosan alvó csoportot (lila)__, illetve a kevesebbet játszók két táborát, akik vagy __sokat (zöld)__, vagy __keveset (sárga) alszanak__. 
A piros X-szel jelölt középpontok (centroidok) mutatják az egyes klaszterek tipikus, "átlagos" képviselőjének jellemzőit.

---

```python

dbscan = DBSCAN(eps=0.4, min_samples=3)
dbscan_labels = dbscan.fit_predict(X_scaled)


plt.figure(figsize=(8, 6))


palette = sns.color_palette("Set2", len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0))
if -1 in dbscan_labels:
    palette.insert(0, (0.0, 0.0, 0.0)) 

sns.scatterplot(x=X['daily_gaming_hours'], y=X['sleep_hours'], hue=dbscan_labels, palette=palette, s=60)

plt.title('DBSCAN Klaszterezés: Játékidő vs. Alvásidő')
plt.xlabel('Napi játékidő (óra)')
plt.ylabel('Alvásidő (óra)')
plt.legend(title='Klaszter (-1 = Zaj)')
plt.show()
```

A DBSCANNÉL 3 nagy csoportot láthatunk:

__0-s klaszter (Zöld):__ Az adatok túlnyomó része egyetlen nagy, összefüggő, sűrű halmazt alkot. Számukra a játékidő és az alvásidő a normál tartományokon belül (0-10 óra játék, 4-10 óra alvás) folyamatos átmenetet képez, nem válnak élesen különálló csoportokra.

__1-es klaszter (Narancssárga):__ Egy apró ponthalmaz, a napi 12-14 órát játszó, nagyjából 7-8 órát alvó játékosoknál.

__-1-es klaszter (Fekete pontok):__ Zaj / Kiemelkedő esetek extrém eseteket, amelyeket nem kényszerített bele egyik klaszterbe sem. Ide tartoznak a legszélsőségesebb adatok (pl. napi 15+ óra játék, vagy extrém kevés alvás magas játékidővel párosítva).

------------------

Mindkét algoritmus ugyanazt a  trendet erősítette meg: a rengeteg játék nem feltétlenül vezet alvásmegvonáshoz.
A K-Means centroidja nagyjából 7 óra alvásnál helyezkedik el.
A DBSCAN talált egy kis pontot, akik sokat játszanak 12-14 óra, de mégis alszanak 7-8 órát.

# B1.) Attribútumok kiválasztása és a feladat megfogalmazása
__A regressziós feladat:__
Cél a játékosok tanulmányi teljesítményének előrejelzése a mindennapi szokásaik (játékidő, alvás) és pszichológiai állapotuk (stressz) alapján.

__Kiválasztott attribútumok:__

__Célváltozó (y):__ academic_performance (Folytonos változó, amit becsülni kell).

__1. modell magyarázó változója (X):__ daily_gaming_hours (Egyszerű lineáris regresszió).

__2. modell magyarázó változói (X):__ daily_gaming_hours, sleep_hours és stress_level (Többszörös lineáris regresszió).

__A kiválasztás indoklása:__
Az A3. rész korrelációs vizsgálatai alapján __ezen változók között összefüggés feltételezhető.__ 
    A megnövekedett napi játékidő logikusan csökkenti a tanulásra fordítható időt. 
    A második modellnél a stressz-szintet és az alvásidőt is bevontuk, mivel a teljesítményt a valóságban is több tényező együttesen 
    befolyásolja, így pontosabb predikciót kaphatunk. 
    A választott változók mind folytonosak, ami a lineáris regresszió alapfeltétele.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python

X1 = df[['daily_gaming_hours']].dropna()
y1 = df['academic_performance'].loc[X1.index]


X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=42)


model1 = LinearRegression()
model1.fit(X_train1, y_train1)


y_pred1 = model1.predict(X_test1)


r2_1 = r2_score(y_test1, y_pred1)
mse_1 = mean_squared_error(y_test1, y_pred1)

print("--- 1. MODELL EREDMÉNYEI ---")
print(f"R-négyzet (Determinációs együttható): {r2_1:.4f}")
print(f"MSE (Átlagos négyzetes hiba): {mse_1:.4f}")


plt.figure(figsize=(8, 5))
plt.scatter(X_test1, y_test1, color='blue', alpha=0.5, label='Valós adatok')
plt.plot(X_test1, y_pred1, color='red', linewidth=2, label='Regressziós egyenes (Predikció)')
plt.title('1. Modell: Játékidő vs. Akadémiai teljesítmény')
plt.xlabel('Napi játékidő (óra)')
plt.ylabel('Akadémiai teljesítmény')
plt.legend()
plt.show()
```

A regressziós modell eredményei és a diagram alapján megállapítható, hogy __nincs lineáris összefüggés a napi játékidő és a diákok tanulmányi eredménye között.__

A majdnem tökéletesen vízszintes piros __regressziós egyenes azt mutatja, hogy a játékra fordított órák számának növekedése nem vonja magával a jegyek romlását vagy javulását.__

Ezt az __R-négyzet__ közel nulla, negatív értéke (-0.0111) is megerősíti: ez azt jelenti, hogy a játékidő mint változó az esetek 0%-ában képes megmagyarázni az akadémiai teljesítmény változását. __A modell predikciós képessége gyakorlatilag nulla.__

__MSE (195.1874):__ Az MSE is mutatja a modell pontatlanságát, a valós értékek (kék pontok) nagy szórással helyezkednek el a regressziós egyenes körül.

__Következtetés:__ A vizsgált mintában a sokat játszó diákok ugyanúgy képesek kiemelkedő tanulmányi eredményeket elérni, mint a keveset játszók. A két változó között nincs lineáris kapcsolat, a regressziós egyenes szinte teljesen vízszintes.

Az R-négyzet (R-squared) értéke továbbra is negatív (-0.0070), ami azt jelenti, hogy __a modell predikciós ereje érdemben nem javult az egyváltozós modellhez képest, a tanulmányi jegyek szórását ezek a változók nem magyarázzák meg.__

Mivel a célváltozó pontosan két lehetséges értéket vehet fel (használ / nem használ), ez egy klasszikus bináris osztályozási probléma, amire a logisztikus regresszió a legalkalmasabb.

__Kiválasztott attribútumok:__

__Célváltozó (y):__ headset_usage (Diszkrét, bináris változó: 0 = Nem, 1 = Igen).

__1. modell magyarázó változója (X):__ competitive_rank (Egyszerű logisztikus regresszió).

__2. modell magyarázó változói (X):__ competitive_rank, daily_gaming_hours és multiplayer_ratio (Többváltozós logisztikus regresszió).

__A kiválasztás indoklása:__
Az A3. részben elkészített dobozábra (boxplot) rámutatott, hogy a kompetitív rang (competitive_rank) eloszlása eltér a headsetet használók és nem használók körében. __Feltételezés, hogy a magasabb rangú, naponta többet játszó (daily_gaming_hours), és a többjátékos módot preferáló (multiplayer_ratio) játékosok nagyobb eséllyel használnak headsetet a csapattársakkal való kommunikáció miatt.__ Ezen magyarázó változók bevonásával a modellünk várhatóan nagy pontossággal tudja majd szétválasztani a két osztályt.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python

X1 = df[['competitive_rank']].dropna()
y1 = df['headset_usage'].loc[X1.index]


X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=42)


log_reg1 = LogisticRegression()
log_reg1.fit(X_train1, y_train1)


y_pred1 = log_reg1.predict(X_test1)


acc1 = accuracy_score(y_test1, y_pred1)
print("--- 1. MODELL EREDMÉNYEI ---")
print(f"Pontosság (Accuracy): {acc1:.4f}")


plt.figure(figsize=(5, 4))
sns.heatmap(confusion_matrix(y_test1, y_pred1), annot=True, fmt='d', cmap='Blues')
plt.title('1. Modell Konfúziós Mátrix')
plt.xlabel('Prediktált érték (0 = Nem, 1 = Igen)')
plt.ylabel('Valós érték')
plt.show()
```

A modell eredményei és a konfúziós mátrix alapján egyértelmű, hogy ez __az egyváltozós modell (amely csak a kompetitív rangot vizsgálta) nem képes megjósolni a headset használatát.__

A pontosság (Accuracy) mindössze 0.4826 (kb. 48%), ami azt jelenti, hogy __a modell teljesítménye rosszabb, mint egy véletlenszerű pénzfeldobásé__ (ami 50% lenne).

__A konfúziós mátrix is ezt tükrözi:__ a modell sokszor téved. __60__- esetben tippelte azt, hogy használ headsetet, pedig nem __44__ esetben tippelte azt, hogy nem használ, pedig igen. 

A hibák száma (104) meghaladja a helyes találatokét (97)

# D

# D1.) Attribútumok kiválasztása és a feladat megfogalmazása
__Az osztályozási feladat:__
Továbbra is a headset_usage (headset használata) bináris (0 vagy 1) célváltozót kell megjósolni, de most  több és változatosabb magyarázó változót kell bevonni. Cél annak vizsgálata, hogy a fejlettebb, nem-lineáris gépi tanulási algoritmusok képesek-e rejtett összefüggéseket találni a játékosok profiljában.

__Kiválasztott attribútumok:__

__Célváltozó (y):__ headset_usage

__Magyarázó változók (X):__ competitive_rank, daily_gaming_hours, multiplayer_ratio, age, stress_level, online_friends.

__Indoklás:__
A C.) részben láttuk, hogy a játékidő és a rang jó prediktorok. __Most kiegészül kor (age), pszichológiai (stress_level) és szociális (online_friends) tényezőkkel.__
Döntési fát és Tartóvektor Gépet (SVM) alkalmazok.


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


features_d = ['competitive_rank', 'daily_gaming_hours', 'multiplayer_ratio', 'age', 'stress_level', 'online_friends']
X_d = df[features_d].dropna()
y_d = df['headset_usage'].loc[X_d.index]


X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(X_d, y_d, test_size=0.2, random_state=42)


scaler_d = StandardScaler()

X_train_scaled = scaler_d.fit_transform(X_train_d)

X_test_scaled = scaler_d.transform(X_test_d)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC



tree_model = DecisionTreeClassifier(max_depth=5, random_state=42)
tree_model.fit(X_train_scaled, y_train_d)

svm_model = SVC(kernel='linear', C=1.0, random_state=42)
svm_model.fit(X_train_scaled, y_train_d)




tree_importance = pd.DataFrame({
    'Attribútum': features_d,
    'Fontosság (Feature Importance)': tree_model.feature_importances_
}).sort_values(by='Fontosság (Feature Importance)', ascending=False)

print("\nDöntési fa attribútumainak fontossága:")
display(tree_importance)

```

A táblázat megmutatja, hogy a Döntési fa (Decision Tree) algoritmus mely változókra támaszkodott a leginkább a headset használatának megjóslása során.

A döntéshozatal során __a kompetitív rang (competitive_rank - 30.2%) és az online barátok száma (online_friends - 26.9%) volt a két legmeghatározóbb tényező. Ez a két attribútum felel az algoritmus döntéseinek több mint feléért.__

Közepesen fontos szempont volt a napi játékidő (daily_gaming_hours) és az életkor (age).

 __A többjátékos hajlandóság__ (multiplayer_ratio) alig 2.7%-kal, __a stresszszint__ pedig mindössze 8.7%-kal járult hozzá a predikcióhoz, tehát ezek a tényezők __szinte egyáltalán nem befolyásolják a headset használatot ebben a nem-lineáris modellben.__

```python

sns.set_theme(style="whitegrid")


plt.figure(figsize=(10, 6))

sns.barplot(
    x='Fontosság (Feature Importance)', 
    y='Attribútum', 
    data=tree_importance, 
    palette='viridis' 
)

plt.title('Döntési fa attribútumainak fontossága', fontsize=14, fontweight='bold')
plt.xlabel('Fontosság (Az értékek összege 1.0)', fontsize=12)
plt.ylabel('Attribútum', fontsize=12)


plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

```

Az ábra mutatja, hogy __a döntési fa predikcióit leginkább a kompetitív rang (competitive_rank) és az online barátok száma (online_friends) határozta meg__. Ezzel szemben __a többjátékos arány (multiplayer_ratio) és a stresszszint (stress_level) szinte alig játszottak szerepet az algoritmus döntéshozatalában__.

```python

y_pred_tree = tree_model.predict(X_test_scaled)
y_pred_svm = svm_model.predict(X_test_scaled)


cm_tree = confusion_matrix(y_test_d, y_pred_tree)
cm_svm = confusion_matrix(y_test_d, y_pred_svm)


fig, axes = plt.subplots(1, 2, figsize=(12, 5))


sns.heatmap(
    cm_tree, annot=True, fmt='d', cmap='Greens', 
    cbar=False, ax=axes[0], annot_kws={'size': 14, 'weight': 'bold'}
)
axes[0].set_title('Konfúziós Mátrix - Döntési Fa', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Prediktált érték', fontsize=12)
axes[0].set_ylabel('Valós érték', fontsize=12)
axes[0].set_xticks([0.5, 1.5])
axes[0].set_xticklabels(['0', '1'])
axes[0].set_yticks([0.5, 1.5])
axes[0].set_yticklabels(['0', '1'], rotation=0)


sns.heatmap(
    cm_svm, annot=True, fmt='d', cmap='Blues', 
    cbar=False, ax=axes[1], annot_kws={'size': 14, 'weight': 'bold'}
)
axes[1].set_title('Konfúziós Mátrix - Támogató Vektor Gép (SVM)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Prediktált érték', fontsize=12)
axes[1].set_ylabel('Valós érték', fontsize=12)
axes[1].set_xticks([0.5, 1.5])
axes[1].set_xticklabels(['0', '1'])
axes[1].set_yticks([0.5, 1.5])
axes[1].set_yticklabels(['0', '1'], rotation=0)

plt.tight_layout()
plt.show()
```

Az __SVM__ modell nyújtotta a legjobb összesített teljesítményt, 56.72%-os pontossággal.
A konfúziós mátrix  alapján ez a modell megbízhatóbban ismeri fel a headsetet nem használókat (60 helyes találat), miközben a használóknál (54 helyes találat) is viszonylag stabil marad.
Az összes aggregált metrikája (Precision, Recall, F1-score) kiegyensúlyozottan 0.57 körül mozog.

A __döntési fa__  predikciója gyengébb, 53.73%-os pontosságot ért el.
A konfúziós mátrixa  és a Classification Report rávilágít a modell gyenge pontjára: rendkívül alacsony a visszahívása (Recall = 0.36) a "headsetet használók" (1-es osztály) esetében. A valóságban __headsetet használók jelentős részét (61 fő) tévesen a "nem használók" kategóriájába sorolta.__ Ez arra utal, hogy a modell túlságosan rátanult a 0-s osztály predikciójára.

```python
from sklearn.metrics import classification_report, accuracy_score


y_pred_tree = tree_model.predict(X_test_scaled)
y_pred_svm = svm_model.predict(X_test_scaled)

print("="*40)
print("DÖNTÉSI FA (DECISION TREE) KIÉRTÉKELÉSE")
print("="*40)
print(f"Pontosság (Accuracy): {accuracy_score(y_test_d, y_pred_tree):.4f}")
print("Részletes jelentés:")
print(classification_report(y_test_d, y_pred_tree))

print("\n" + "="*40)
print("TÁMOGATÓ VEKTOR GÉP (SVM) KIÉRTÉKELÉSE")
print("="*40)
print(f"Pontosság (Accuracy): {accuracy_score(y_test_d, y_pred_svm):.4f}")
print("Részletes jelentés:")
print(classification_report(y_test_d, y_pred_svm))



```
