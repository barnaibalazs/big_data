```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN


X = df[['daily_gaming_hours', 'sleep_hours']].dropna()


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python

err = []

for i in range(1, 11):
   
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10) 
    kmeans.fit(X_scaled)
    err.append([kmeans.n_clusters, kmeans.inertia_])


err = np.asarray(err) 


plt.figure(figsize=(8, 5))
sns.lineplot(x=err[:,0], y=err[:,1], marker='o', color='b')
sns.scatterplot(x=err[:,0], y=err[:,1], color='red', s=80, zorder=5)

plt.title('Könyök módszer az optimális klaszterszám (K) meghatározásához')
plt.xlabel('Klaszterek száma (K)')
plt.ylabel('Inercia (Klaszteren belüli hiba)')
plt.xticks(range(1, 11))
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

A görbén látszik, hogy a hiba meredek csökkenése a __K=3__ értéknél törik meg és kezd ellaposodni. 
Ez tökéletesen igazolja, hogy  K-Means algoritmusnál a __3 klaszter__ (játékosprofil) beállítása  az optimális választás, hiszen ennél több csoport kialakítása már nem csökkentené érdemben a hibát.

```python

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_scaled)


plt.figure(figsize=(8, 6))

sns.scatterplot(x=X['daily_gaming_hours'], y=X['sleep_hours'], hue=kmeans_labels, palette='viridis', s=60)


centers = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroidok')

plt.title('K-Means Klaszterezés (k=3): Játékidő vs. Alvásidő')
plt.xlabel('Napi játékidő (óra)')
plt.ylabel('Alvásidő (óra)')
plt.legend(title='Klaszter')
plt.show()
```

A K-Means algoritmus három jól értelmezhető játékosprofilt különített el: a sokat játszó, átlagosan alvó csoportot (lila), illetve a kevesebbet játszók két táborát, akik vagy sokat (zöld), vagy keveset (sárga) alszanak. 
A piros X-szel jelölt középpontok (centroidok) tökéletesen mutatják az egyes klaszterek tipikus, "átlagos" képviselőjének jellemzőit.

---

```python

dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)


plt.figure(figsize=(8, 6))


palette = sns.color_palette("Set2", len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0))
if -1 in dbscan_labels:
    palette.insert(0, (0.0, 0.0, 0.0)) # Fekete szín a zajnak

sns.scatterplot(x=X['daily_gaming_hours'], y=X['sleep_hours'], hue=dbscan_labels, palette=palette, s=60)

plt.title('DBSCAN Klaszterezés: Játékidő vs. Alvásidő')
plt.xlabel('Napi játékidő (óra)')
plt.ylabel('Alvásidő (óra)')
plt.legend(title='Klaszter (-1 = Zaj)')
plt.show()
```

A DBSCAN algoritmus az adatok túlnyomó többségét egyetlen nagy, sűrű klaszterbe (zöld) vonta össze, ami a játékosok tipikus, átlagos viselkedését reprezentálja. 
A széleken elszórtan elhelyezkedő fekete pontok (-1-es zaj) tökéletesen azonosítják a kilógó eseteket, vagyis azokat az extrém felhasználókat, akik például abnormálisan sokat játszanak vagy szélsőségesen keveset alszanak.

# B1.) Attribútumok kiválasztása és a feladat megfogalmazása
__A regressziós feladat:__
Cél a játékosok tanulmányi teljesítményének előrejelzése a mindennapi szokásaik (játékidő, alvás) és pszichológiai állapotuk (stressz) alapján.

__Kiválasztott attribútumok:__

__Célváltozó (y):__ academic_performance (Folytonos változó, amit becsülni kell).

__1. modell magyarázó változója (X):__ daily_gaming_hours (Egyszerű lineáris regresszió).

__2. modell magyarázó változói (X):__ daily_gaming_hours, sleep_hours és stress_level (Többszörös lineáris regresszió).

__A kiválasztás indoklása:__
Az A3. rész korrelációs vizsgálatai alapján ezen változók között összefüggés feltételezhető. 
    A megnövekedett napi játékidő logikusan csökkenti a tanulásra fordítható időt. 
    A második modellnél a stressz-szintet és az alvásidőt is bevontuk, mivel a teljesítményt a valóságban is több tényező együttesen 
    befolyásolja, így pontosabb predikciót kaphatunk. 
    A választott változók mind folytonosak, ami a lineáris regresszió alapfeltétele.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python

X1 = df[['daily_gaming_hours']].dropna()
y1 = df['academic_performance'].loc[X1.index]


X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=42)


model1 = LinearRegression()
model1.fit(X_train1, y_train1)


y_pred1 = model1.predict(X_test1)


r2_1 = r2_score(y_test1, y_pred1)
mse_1 = mean_squared_error(y_test1, y_pred1)

print("--- 1. MODELL EREDMÉNYEI ---")
print(f"R-négyzet (Determinációs együttható): {r2_1:.4f}")
print(f"MSE (Átlagos négyzetes hiba): {mse_1:.4f}")


plt.figure(figsize=(8, 5))
plt.scatter(X_test1, y_test1, color='blue', alpha=0.5, label='Valós adatok')
plt.plot(X_test1, y_pred1, color='red', linewidth=2, label='Regressziós egyenes (Predikció)')
plt.title('1. Modell: Játékidő vs. Akadémiai teljesítmény')
plt.xlabel('Napi játékidő (óra)')
plt.ylabel('Akadémiai teljesítmény')
plt.legend()
plt.show()
```

A regressziós modell eredményei és a diagram alapján megállapítható, hogy __nincs lineáris összefüggés a napi játékidő és a diákok tanulmányi eredménye között.__

A majdnem tökéletesen vízszintes piros __regressziós egyenes azt mutatja, hogy a játékra fordított órák számának növekedése nem vonja magával a jegyek romlását vagy javulását.__

Ezt az R-négyzet (R-squared) közel nulla, sőt negatív értéke (-0.0111) is megerősíti: ez azt jelenti, hogy a játékidő mint változó az esetek 0%-ában képes megmagyarázni az akadémiai teljesítmény változását. A modell predikciós képessége gyakorlatilag nulla.

---

```python

features = ['daily_gaming_hours', 'sleep_hours', 'stress_level']


X2 = df[features].dropna()
y2 = df['academic_performance'].loc[X2.index]


X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2, random_state=42)


model2 = LinearRegression()
model2.fit(X_train2, y_train2)


y_pred2 = model2.predict(X_test2)
r2_2 = r2_score(y_test2, y_pred2)
mse_2 = mean_squared_error(y_test2, y_pred2)

print("--- 2. MODELL EREDMÉNYEI ---")
print(f"R-négyzet (Determinációs együttható): {r2_2:.4f}")
print(f"MSE (Átlagos négyzetes hiba): {mse_2:.4f}")


koefficiens_df = pd.DataFrame({
    'Attribútum': features, 
    'Együttható (Súly)': model2.coef_
})
print("\nA változók hatása a tanulmányi teljesítményre:")
display(koefficiens_df)
```

A többváltozós regressziós modell eredményei rámutatnak, hogy __a napi játékidő, az alvásmennyiség és a stresszszint együttesen sem képesek lineárisan megmagyarázni a tanulmányi eredmények alakulását.__

Az együtthatók (súlyok) irányát megvizsgálva: az alvásidő (sleep_hours) együtthatója pozitív (0.39), tehát a több alvás a modell szerint enyhén felfelé húzza a teljesítményt. 
Ezzel szemben a napi játékidő (daily_gaming_hours) együtthatója negatív (-0.25), ami nagyon enyhe rontó hatást feltételez.

# D

# D1.) Attribútumok kiválasztása és a feladat megfogalmazása
__Az osztályozási feladat:__
Továbbra is a headset_usage (headset használata) bináris (0 vagy 1) célváltozót kell megjósolni, de most sokkal több és változatosabb magyarázó változót kell bevonni. Cél annak vizsgálata, hogy a fejlettebb, nem-lineáris gépi tanulási algoritmusok képesek-e rejtett összefüggéseket találni a játékosok profiljában.

__Kiválasztott attribútumok:__

__Célváltozó (y):__ headset_usage

__Magyarázó változók (X):__ competitive_rank, daily_gaming_hours, multiplayer_ratio, age, stress_level, online_friends.

__Indoklás:__
A C.) részben láttuk, hogy a játékidő és a rang jó prediktorok. Most kiegészítjük ezt demográfiai (age), pszichológiai (stress_level) és szociális (online_friends) tényezőkkel. Egy Döntési fa vagy egy Támogató Vektor Gép (SVM) képes a változók közötti komplex, többszörös interakciókat is leképezni (pl. lehet, hogy a fiatalabbak kevesebb játéknál is használnak headsetet, míg az idősebbek csak kompetitív szinten).

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


features_d = ['competitive_rank', 'daily_gaming_hours', 'multiplayer_ratio', 'age', 'stress_level', 'online_friends']
X_d = df[features_d].dropna()
y_d = df['headset_usage'].loc[X_d.index]


X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(X_d, y_d, test_size=0.2, random_state=42)


scaler_d = StandardScaler()

X_train_scaled = scaler_d.fit_transform(X_train_d)

X_test_scaled = scaler_d.transform(X_test_d)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC



tree_model = DecisionTreeClassifier(max_depth=5, random_state=42)
tree_model.fit(X_train_scaled, y_train_d)

svm_model = SVC(kernel='linear', C=1.0, random_state=42)
svm_model.fit(X_train_scaled, y_train_d)

print("Modellek sikeresen betanítva!")


tree_importance = pd.DataFrame({
    'Attribútum': features_d,
    'Fontosság (Feature Importance)': tree_model.feature_importances_
}).sort_values(by='Fontosság (Feature Importance)', ascending=False)

print("\nDöntési fa attribútumainak fontossága:")
display(tree_importance)
```

A táblázat megmutatja, hogy a Döntési fa (Decision Tree) algoritmus mely változókra támaszkodott a leginkább a headset használatának megjóslása során (az értékek összege 1, azaz 100%).

A döntéshozatal során egyértelműen __a kompetitív rang (competitive_rank - 30.2%) és az online barátok száma (online_friends - 26.9%) volt a két legmeghatározóbb tényező. Ez a két attribútum felel az algoritmus döntéseinek több mint feléért.__

Közepesen fontos szempont volt a napi játékidő (daily_gaming_hours) és az életkor (age).

Meglepő módon __a többjátékos hajlandóság__ (multiplayer_ratio) alig 2.7%-kal, __a stresszszint__ pedig mindössze 8.7%-kal járult hozzá a predikcióhoz, tehát ezek a tényezők __szinte egyáltalán nem befolyásolják a headset használatot ebben a nem-lineáris modellben.__

---

```python
from sklearn.metrics import classification_report, accuracy_score


y_pred_tree = tree_model.predict(X_test_scaled)
y_pred_svm = svm_model.predict(X_test_scaled)

print("="*40)
print("DÖNTÉSI FA (DECISION TREE) KIÉRTÉKELÉSE")
print("="*40)
print(f"Pontosság (Accuracy): {accuracy_score(y_test_d, y_pred_tree):.4f}")
print("Részletes jelentés:")
print(classification_report(y_test_d, y_pred_tree))

print("\n" + "="*40)
print("TÁMOGATÓ VEKTOR GÉP (SVM) KIÉRTÉKELÉSE")
print("="*40)
print(f"Pontosság (Accuracy): {accuracy_score(y_test_d, y_pred_svm):.4f}")
print("Részletes jelentés:")
print(classification_report(y_test_d, y_pred_svm))
```

A fejlettebb, __nem-lineáris gépi tanulási algoritmusok bevetése javított a korábbi logisztikus regresszió (48-49%) eredményein__, de a modellek predikciós ereje még mindig korlátozott.

A Döntési fa pontossága (Accuracy) 0.5373 (kb. 54%). A részletes jelentésből kiolvasható, hogy a modell erősen részrehajló: __viszonylag jól felismeri azokat a játékosokat, akik nem használnak headsetet__ (0-s osztály, Recall: 70%), __viszont nagyon sokszor elvéti azokat, akik igen__ (1-es osztály, Recall: mindössze 36%). __Emiatt a modell F1-score értékei is nagyon aszimmetrikusak__ (0.61 vs 0.42).

__A Támogató Vektor Gép (SVM) teljesített a legjobban a projekt során épített modellek közül__, 0.5672-es (kb. 57%) pontossággal. Ami ennél is fontosabb: __az SVM sokkal kiegyensúlyozottabb döntéseket hoz.__ 
A visszahívás (Recall) mindkét osztály (0 és 1) esetében pontosan 57%, ami azt mutatja, hogy az algoritmus nem húz egyik kategória felé sem, stabilabb és megbízhatóbb alapot ad.
